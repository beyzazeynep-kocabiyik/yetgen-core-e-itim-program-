# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self):
        self.tv_acik = False
        self.ses_seviyesi = 10
        self.kanallar = []
        self.aktif_kanal = None

    def tv_ac(self):
        if not self.tv_acik:
            self.tv_acik = True
            print("Televizyon açıldı.")
        else:
            print("Televizyon zaten açık.")

    def tv_kapat(self):
        if self.tv_acik:
            self.tv_acik = False
            print("Televizyon kapatıldı.")
        else:
            print("Televizyon zaten kapalı.")

    def ses_ayarlari(self):
        if not self.tv_acik:
            print("Televizyon kapalı. Ses ayarlarını yapamazsınız.")
            return
        try:
            yeni_ses = int(input("Yeni ses seviyesini girin (0-100): "))
            if 0 <= yeni_ses <= 100:
                self.ses_seviyesi = yeni_ses
                print(f"Ses seviyesi {self.ses_seviyesi} olarak ayarlandı.")
            else:
                print("Ses seviyesi 0 ile 100 arasında olmalıdır.")
        except ValueError:
            print("Geçersiz giriş. Lütfen bir sayı girin.")

    def kanal_ekle(self):
        if not self.tv_acik:
            print("Televizyon kapalı. Kanal ekleyemezsiniz.")
            return
        kanal_adı = input("Eklemek istediğiniz kanal adını girin: ")
        if kanal_adı not in self.kanallar:
            self.kanallar.append(kanal_adı)
            print(f"Kanal '{kanal_adı}' eklendi.")
        else:
            print("Kanal zaten mevcut.")

    def kanal_sayisini_ogrenme(self):
        if not self.tv_acik:
            print("Televizyon kapalı. Kanal sayısını öğrenemezsiniz.")
            return
        print(f"Toplam {len(self.kanallar)} kanal mevcut.")

    def rastgele_kanala_gec(self):
        if not self.tv_acik:
            print("Televizyon kapalı. Rastgele kanala geçemezsiniz.")
            return
        if not self.kanallar:
            print("Kanal listesi boş.")
            return
        self.aktif_kanal = random.choice(self.kanallar)
        print(f"Rastgele olarak '{self.aktif_kanal}' kanalına geçildi.")

    def tv_bilgileri(self):
        if not self.tv_acik:
            print("Televizyon kapalı. Bilgileri gösteremezsiniz.")
            return
        print(f"Televizyon {'açık' if self.tv_acik else 'kapalı'}.")
        print(f"Ses seviyesi: {self.ses_seviyesi}")
        print(f"Mevcut kanallar: {', '.join(self.kanallar) if self.kanallar else 'Kanal bulunmuyor.'}")
        print(f"Aktif kanal: {self.aktif_kanal if self.aktif_kanal else 'Aktif kanal yok.'}")

def kumanda_yonetimi():
    kumanda = Kumanda()
    while True:
        print("\nYapmak istediğiniz işlemi seçin:")
        print("1. TV Aç")
        print("2. TV Kapat")
        print("3. Ses Ayarları")
        print("4. Kanal Ekle")
        print("5. Kanal Sayısını Öğrenme")
        print("6. Rastgele Kanala Geçme")
        print("7. Televizyon Bilgileri")
        print("q. Çıkış")
        
        secim = input("Seçiminizi yapın: ")
        
        if secim == "1":
            kumanda.tv_ac()
        elif secim == "2":
            kumanda.tv_kapat()
        elif secim == "3":
            kumanda.ses_ayarlari()
        elif secim == "4":
            kumanda.kanal_ekle()
        elif secim == "5":
            kumanda.kanal_sayisini_ogrenme()
        elif secim == "6":
            kumanda.rastgele_kanala_gec()
        elif secim == "7":
            kumanda.tv_bilgileri()
        elif secim.lower() == "q":
            print("Kumanda kapatılıyor.")
            break
        else:
            print("Geçersiz seçim. Lütfen tekrar deneyin.")
        
        time.sleep(1)  

if __name__ == "__main__":
    kumanda_yonetimi()
```
