# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python

class Vehicle:
    def __init__(self, name):
        self.name = name


class Bus(Vehicle):
    def __init__(self, name, seating_capacity=50):
        super().__init__(name)  
        self.seating_capacity = seating_capacity

    def seating_capacity(self):
        
        return f"The seating capacity of a bus is {self.seating_capacity} passengers"


my_bus = Bus("School Bus")


print(my_bus.seating_capacity())

````

# Problem 2


School_bus'ın aynı zamanda Vehicle sınıfının bir örneği olup olmadığını belirleyiniz.

````python

class Vehicle:
    def __init__(self, name):
        self.name = name

class Bus(Vehicle):
    def __init__(self, name, seating_capacity=50):
        super().__init__(name)  
        self.seating_capacity = seating_capacity

    def seating_capacity(self):
        """
        Otobüsün oturma kapasitesini döndürür.
        """
        return f"The seating capacity of a bus is {self.seating_capacity} passengers"

school_bus = Bus("School Bus")


if isinstance(school_bus, Vehicle):
    print("school_bus is an instance of Vehicle.")
else:
    print("school_bus is not an instance of Vehicle.")
````
