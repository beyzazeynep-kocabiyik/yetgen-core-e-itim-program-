# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def urun_kaydet():
    """
    Kullanıcıdan ürün adı ve fiyatını alır ve bu bilgileri 'urunler.txt' dosyasına kaydeder.
    """
    urun_ad = input("Ürün adını girin: ")
    while True:
        try:
            urun_fiyat = float(input("Ürün fiyatını girin: "))
            break  
        except ValueError:
            print("Lütfen geçerli bir fiyat girin.")
    
    with open("urunler.txt", "a") as dosya:
        dosya.write(f"Ürün Adı: {urun_ad}, Ürün Fiyatı: {urun_fiyat:.2f}\n")
    
    print("Ürün bilgileri başarıyla kaydedildi.")

urun_kaydet()
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
def kelime_degistir():
    
    eski_kelime = input("Eski kelimeyi girin: ")
    yeni_kelime = input("Yeni kelimeyi girin: ")
    
    try:
       
        with open("urunler.txt", "r") as dosya:
            icerik = dosya.read()
        
       
        yeni_icerik = icerik.replace(eski_kelime, yeni_kelime)
        
        
        with open("urunler.txt", "w") as dosya:
            dosya.write(yeni_icerik)
        
        print("Kelime başarıyla değiştirildi.")
    except FileNotFoundError:
        print("Dosya bulunamadı. Lütfen 'urunler.txt' dosyasının mevcut olduğundan emin olun.")
    except Exception as e:
        print(f"Bir hata oluştu: {e}")


kelime_degistir()
   
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python

class Vehicle:
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage

    def display_info(self):
        return f"Name: {self.name} Speed: {self.speed} Mileage: {self.mileage}"

class Bus(Vehicle):
    def __init__(self, name, speed, mileage):
       
        super().__init__(name, speed, mileage)


my_bus = Bus("School Volvo", 180, 12)


print(my_bus.display_info())
````
