# Problem 1

Fizikte, bir nesnenin sabit ivmeyle hareket ederken son hızını 
(veya hızını) bulmak için aşağıdaki denklem kullanılabilir:
```
vf = vi + at
burada:
vf= son hız
vi= ilk hız
a= hızlanma
t= zaman
```
İlk hız, ivme ve zaman verildiğinde, son hızı döndürecek bir fonksiyon yazın.

````python
def son_hiz(vi, a, t):
    """
    Bu fonksiyon verilen ilk hız, ivme ve zamanı kullanarak son hızı hesaplar.
    
    :param vi: İlk hız (m/s)
    :param a: İvme (m/s²)
    :param t: Zaman (saniye)
    :return: Son hız (m/s)
    """
    vf = vi + a * t
    return vf
ilk_hiz = 10  # m/s
ivme = 2  # m/s²
zaman = 5  # saniye

print(f"Son hız: {son_hiz(ilk_hiz, ivme, zaman)} m/s")
  
````

# Problem 2

1'den 1000'e kadar olan sayılardan mükemmel sayı olanları ekrana yazdırın. Bunun için bir sayının mükemmel olup olmadığını dönen bir tane fonksiyon yazın.

Bir sayının bölenlerinin toplamı kendine eşitse bu sayı mükemmel bir sayıdır. Örnek olarak 6 mükemmel bir sayıdır (1 + 2 + 3 = 6).

````python
def mukemmel_mi(sayi):
    """
    Bu fonksiyon verilen bir sayının mükemmel sayı olup olmadığını kontrol eder.
    
    :param sayi: Kontrol edilecek sayı
    :return: Eğer sayı mükemmel ise True, değilse False
    """
    toplam = 0
    for i in range(1, sayi):
        if sayi % i == 0:
            toplam += i
    return toplam == sayi

def mukemmel_sayilari_bul():
    """
    1'den 1000'e kadar olan mükemmel sayıları bulur ve ekrana yazdırır.
    """
    for sayi in range(1, 1001):
        if mukemmel_mi(sayi):
            print(sayi)

# Fonksiyonu çağırarak mükemmel sayıları bulalım
mukemmel_sayilari_bul()

````

# Problem 4

Bir duvar boyamaya karar verdiniz. Boya kutusunun üzerindeki talimatta, 1 kutu boyanın 5 metrekarelik bir duvarı boyayabileceği yazıyor. Rastgele bir duvar yüksekliği ve genişliği verildiğinde, kaç kutu boya satın almanız gerektiğini hesaplayın.

kutu sayısı = (duvar yüksekliği * duvar genişliği) ÷ kutu başına kaplama.

örneğin Yükseklik = 2, Genişlik = 4, Kaplama = 5

kutu sayısı = (2 * 4) ÷ 5 = 1.6

Ancak bir kutu boyanın 0,6'sını satın alamayacağınız için, sonuç 2 kutuya yuvarlanmalıdır .


````python
import math

def kutu_sayisi(yukseklik, genislik, kaplama):
    """
    Bu fonksiyon verilen duvar yüksekliği, genişliği ve kutu başına kaplama bilgilerini kullanarak
    gerekli kutu sayısını hesaplar ve en yakın üst tam sayıya yuvarlar.
    
    :param yukseklik: Duvarın yüksekliği (metre)
    :param genislik: Duvarın genişliği (metre)
    :param kaplama: Bir kutunun kaplayabileceği alan (metrekare)
    :return: Gerekli kutu sayısı (tam sayı)
    """
    alan = yukseklik * genislik
    gerekli_kutu = alan / kaplama
    return math.ceil(gerekli_kutu)

# Örnek kullanım
yukseklik = 2
genislik = 4
kaplama = 5

print(f"Gerekli kutu sayısı: {kutu_sayisi(yukseklik, genislik, kaplama)}")
````

# Problem 6

Verilen bir listeden çift sayıları yazdıran bir fonksiyon yazınız.


````python
def cift_sayilari_yazdir(liste):
    """
    Bu fonksiyon verilen bir listeden çift sayıları yazdırır.
    
    :param liste: Sayıları içeren liste
    """
    cift_sayilar = [sayi for sayi in liste if sayi % 2 == 0]
    
    for sayi in cift_sayilar:
        print(sayi)

````

# Problem 7

0'dan 10'a kadar olan sayıların toplamını hesaplayan özyinelemeli (Recursive Function) bir fonksiyon oluşturan bir fonksiyon yazınız.

Özyinelemeli (Recursive Function) bir işlev, kendini tekrar tekrar çağıran bir işlevdir.

````python
def toplam(n):
    
    if n == 0:
        return 0

    else:
        return n + toplam(n - 1)


sonuc = toplam(10)
print("0'dan 10'a kadar olan sayıların toplamı:", sonuc)
````
